```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_csv('pnad.csv')
```

```python
tamanho = dados.shape[0]

data = pd.DataFrame()
```

```python
#Variavel Sexo/ Qualitativa
data["Sexo"] = dados.V0302.astype("category")
data.Sexo.cat.categories = ("Masculino", "Feminino")
```

```python
#Variavel Cor ou Raça/ Qualitativa
data["Cor_ou_raca"] = dados.V0404.astype("category")
data.Cor_ou_raca.cat.categories = ("Indigena", "Branco","Preta","Amarela","Pardo","Sem Declaração")
```

```python
#Variavel Rede de Ensino/ Qualitativa
data["Rede_de_Ensino"] = dados.V6002.astype("category")
data.Rede_de_Ensino.cat.categories = ("Publica", "Privada")
```

```python
#Variavel Nivel de Ensino/ Qualitativa
data["Nivel_de_Ensino"] = dados.V4802.astype("category")
data.Nivel_de_Ensino.cat.categories = ("Maternal, jardim de infância etc.",
"Classe de alfabetização",
"Alfabetização de jovens e adultos",
"Elementar (primário)",
"Médio 1º ciclo (ginasial, etc.)",
"Médio 2º ciclo (científico, clássico, etc.)",
"Regular do ensino fundamental ou do 1º grau",
"Regular do ensino médio ou do 2º grau",
"Educação de jovens e adultos ou supletivo do ensino fundamental ou do 1º grau",
"Educação de jovens e adultos ou supletivo de ensino médio ou do 2º grau",
"Superior de graduação, Mestrado ou doutorado")
```

```python
#Variavel Região/ Qualitativa
data["Regiao"] = dados.UF

data.Regiao[(dados.UF > 10) & (dados.UF < 20)] = "Norte"
data.Regiao[(dados.UF > 20) & (dados.UF < 30)] = "Nordeste"
data.Regiao[(dados.UF > 30) & (dados.UF < 40)] = "Sudeste" 
data.Regiao[(dados.UF > 40) & (dados.UF < 50)] = "Sul" 
data.Regiao[(dados.UF >= 50) & (dados.UF < 60)] = "Centro Oeste" 
```

```python
#Variavel Area de Ensino/ Qualitativa
data["Área_de_Ensino"] = dados.V6020.astype("category")
data.Área_de_Ensino.cat.categories = ("Municipal","Estadual","Federal")
```

```python
#Variavel Creche/ Qualitativa
data["Creche"] = dados.V0606.astype("category")
data.Creche.cat.categories = ("Sim","Não")
```

```python
#Variavel Anos de Estudo/ Quantitativa
data["Anos_De_Estudo"] = dados.V4803.astype("category")
```

```python
#Variavel Renda Mensal/ Quantitativa
data["Rendimento Mensal"] = dados.V4742.astype("category")
```

```python
print("Frequencias relativas por Sexo:")
data1 = ((data.Sexo.value_counts()/tamanho)*100).round(decimals=1)
print(data1, "\n")
plot_graph1 = data1.plot(kind="pie", title="Sexo", autopct="%.1f", figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
```

```python
print("Frequencias de cor ou raca")
data2 = ((data.Cor_ou_raca.value_counts()/tamanho)*100).round(decimals=1)
print(data2, "\n")
plot_graph2 = data2.plot(kind="bar", title="Raça ou Cor", color = ("b","g","k","r","m","c"), figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
plt.xlabel("Cor ou Raça")
```

```python
print("Frequencias relativas por Rede de Ensino:")
data3 = ((data.Rede_de_Ensino.value_counts()/tamanho)*100).round(decimals=1)
print(data3, "\n")

plot_graph3 = data3.plot(kind="pie", title="Rede de Ensino", autopct="%.1f", figsize=(6,6))
```

```python
print("Frequencias relativas por Nivel de Ensino:")
data.Nivel_de_Ensino.cat.categories = ("Maternal, jardim de infância etc.",
"Classe de alfabetização",
"Alfabetização de jovens e adultos",
"Elementar (primário)",
"Médio 1º ciclo (ginasial, etc.)",
"Médio 2º ciclo (científico, clássico, etc.)",
"Regular do ensino fundamental ou do 1º grau",
"Regular do ensino médio ou do 2º grau",
"Educação de jovens e adultos ou supletivo do ensino fundamental ou do 1º grau",
"Educação de jovens e adultos ou supletivo de ensino médio ou do 2º grau",
"Superior de graduação, Mestrado ou doutorado")
data4 = ((data.Nivel_de_Ensino.value_counts()/tamanho)*100).round(decimals=1)
#print(data4, "\n")
plot_graph4 = data4.plot(kind="bar", title="Nivel de Ensino", color = ("b","g","k","r","m","c","y"), figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
plt.xlabel("Nivel de Ensino")
```

```python
data5 = ((data.Regiao.value_counts()/tamanho)*100).round(decimals=1)
print(data5, "\n")
plot_graph5 = data5.plot(kind="pie", title="Região", autopct="%.1f", figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
plt.xlabel("Regiões")
```

```python
print("Frequencias relativas por Area de Ensino:")
data.Área_de_Ensino.cat.categories = ("Municipal","Estadual","Federal")
data6 = ((data.Área_de_Ensino.value_counts()/tamanho)*100).round(decimals=1)
print(data6, "\n")
plot_graph6 = data6.plot(kind="pie", title="Área de Ensino", autopct="%.1f", figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
```

```python
print("Frequencias relativas caso tenha frequentado uma creche ou uma escola:")
data6 = ((data.Creche.value_counts()/tamanho)*100).round(decimals=1)
print(data6, "\n")
plot_graph6 = data6.plot(kind="bar", title="Frequentou Creche ou Escola", color = ("b","g"), figsize=(6,6))
plt.ylabel("Freq. Relativa (em%)")
plt.xlabel("Frequentou Creche ou Escola")
```
